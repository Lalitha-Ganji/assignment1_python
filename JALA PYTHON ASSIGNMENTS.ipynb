```python
#9. Write a program to find the prime or not
# Program to check if a number is prime

def is_prime(n):
    if n < 2:
        return False  # Numbers less than 2 are not prime
    for i in range(2, int(n ** 0.5) + 1):  # Check divisibility up to sqrt(n)
        if n % i == 0:
            return False  # If divisible, it's not prime
    return True  # If no divisors found, it's prime

# Input from user
num = int(input("Enter a number: "))

# Check and display result
if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

```python
#10. Write a function to find the duplicate values of an array
def find_duplicates(arr):
    seen = set()
    duplicates = set()

    for num in arr:
        if num in seen:
            duplicates.add(num)
        else:
            seen.add(num)

    return list(duplicates)  # Returning as a list for better readability

# Example usage:
numbers = [1, 2, 3, 4, 5, 2, 3, 6, 7, 3]
duplicate_numbers = find_duplicates(numbers)
print("Original array:", numbers)
print("Duplicate values:", duplicate_numbers)
```

```python
#1.     Define a static variable and access that through a class
# In Python, we can define a static variable inside a class and access it using the class name. Static variables are shared across all instances of a class and remain unchanged unless explicitly modified.
# Example:
class MyClass:
    static_variable = "I am static!"  # Defining a static variable

# Accessing the static variable through the class
print(MyClass.static_variable)

# Creating an instance and accessing the static variable
instance = MyClass()
print(instance.static_variable)
```

```python
#2. Define a static variable and access that through a instance
# In Python, static variables (class attributes) are shared among all instances of a class. While instances can access them, they should ideally be accessed through the class itself.
# Example:
class MyClass:
    static_variable = "I am a static variable!"

# Creating an instance
instance = MyClass()

# Accessing the static variable through the instance
print(instance.static_variable)
# However, modifying it through an instance creates an instance attribute instead:
instance.static_variable = "New value!"
print(MyClass.static_variable)  # Still unchanged
print(instance.static_variable)  # Changed for this instance only
```

```python
#3.  Define a static variable and change within the instance
# In Python, static variables (class attributes) are shared among all instances of a class. If we modify a static variable within an instance, it creates a new instance attribute instead of modifying the class-level variable.
# Example:
class MyClass:
    static_variable = "I am static!"

# Creating an instance
instance = MyClass()

# Modifying static_variable through the instance
instance.static_variable = "Changed inside instance"

# Checking values
print(MyClass.static_variable)  # Still the original class attribute
print(instance.static_variable)  # Modified instance attribute


# Modifying the Static Variable Properly
# To modify the actual static variable across all instances, use the class name:
instance2 = MyClass()
MyClass.static_variable = "Modified at class level"

print(instance.static_variable)  # Instance-level change persists


print(instance2.static_variable)  # Now reflects the updated class attribute
print(MyClass.static_variable)  # Shows the updated value
```

```python
#4. Define a static variable and change within the class
# In Python, we can define a static variable (class attribute) and modify it at the class level, ensuring that the change applies to all instances. The proper way to do this is by using the class name.
# EXAMPLE:
class MyClass:
    static_variable = "I am static!"

# Modifying the static variable at the class level
MyClass.static_variable = "Changed at class level"

# Creating instances
instance1 = MyClass()
instance2 = MyClass()

# Checking values
print(instance1.static_variable)  # Reflects class-level change
print(instance2.static_variable)  # Reflects class-level change
print(MyClass.static_variable)    # Shows updated value

# Using Class Methods for Controlled Updates
# For better structure, we can modify class attributes using a class method:
class MyClass:
    static_variable = "I am static!"

    @classmethod
    def update_static(cls, new_value):
        cls.static_variable = new_value

# Changing the static variable using the class method
MyClass.update_static("Updated via class method")

print(MyClass.static_variable)  # Shows updated value
print(MyClass().static_variable)  # Instance reflects change
```

```python
#6. Matching a String Against a Regular Expression With matches()
# In Python, we can use the re.fullmatch() function to check if a string exactly matches a given regular expression.
# EXAMPLE:
import re

pattern = r"\d{4}-\d{2}-\d{2}"  # Matches a date format like "2025-05-27"
string = "2025-05-27"

# Using re.fullmatch()
if re.fullmatch(pattern, string):
    print("String matches the pattern!")
else:
    print("String does not match.")
#Explanation:
# The pattern r"\d{4}-\d{2}-\d{2}" matches a date format (YYYY-MM-DD).
# re.fullmatch(pattern, string) returns None if the entire string doesn't match.
# The if condition checks for a successful match.
# Partial Matches (re.search())
# If we want to check if part of the string matches, use re.search() instead:
if re.search(pattern, "Today's date is 2025-05-27"):
    print("Pattern found in the string!")
```

```python
#7. Comparing strings
# Python, we can compare strings using various methods. Here are some common ways:
# 1. Using == (Equality Check)
string1 = "Hello"
string2 = "Hello"
string3 = "World"

print(string1 == string2)  # True (Exact match)
print(string1 == string3)  # False (Different strings)
# 2. Using != (Not Equal)
print(string1 != string3)  # True (Strings are different)
# 3. Using <, >, <=, >= (Lexicographical Comparison)
# Python compares strings character by character based on ASCII values.
print("apple" < "banana")  # True ('a' comes before 'b')
print("Zebra" > "Apple")   # True ('Z' has a higher ASCII value than 'A')
# 4. Using startswith() and endswith()
string = "Python programming"

print(string.startswith("Python"))  # True
print(string.endswith("ming"))  # True
# 5. Using in (Checking Substring Presence)
print("gram" in string)  # True (Substring exists)
# 6. Using lower() or upper() for Case-Insensitive Comparison
string1 = "Hello"
string2 = "hello"

print(string1.lower() == string2.lower())  # True (Converted to lowercase)
```

```python
#A, B and C are classes
#A is a super class. B is a sub class of A. C is a sub class of B.
#Create three methods in each class, 2 methods are specific to each class and third method (override method) should be in all three Classes A, B and C
#Create a class with main method. Create an object for each class A, B and C in main method and call every method of each class using its own object/instance.
#Call an overridden method with super class reference to B and C class’s objects
#Runtime Polymorphism with Data Members/Instance variables, Repeat the above process only for data members
#Python program that demonstrates Inheritance, Method Overriding, and Runtime Polymorphism with Data Members.
#________________________________________
#Step 1: Creating Superclass and Subclasses with Overridden Methods
#Each class has two unique methods and one overridden method.
# Superclass A
class A:
    data = "Data from A"  # Instance variable

    def method1(self):
        print("Method1: Specific to class A")

    def method2(self):
        print("Method2: Specific to class A")

    def overridden_method(self):
        print("Overridden Method: Defined in class A")

# Subclass B (inherits from A)
class B(A):
    data = "Data from B"  # Instance variable

    def method3(self):
        print("Method3: Specific to class B")

    def method4(self):
        print("Method4: Specific to class B")

    def overridden_method(self):
        print("Overridden Method: Overridden in class B")

# Subclass C (inherits from B)
class C(B):
    data = "Data from C"  # Instance variable

    def method5(self):
        print("Method5: Specific to class C")

    def method6(self):
        print("Method6: Specific to class C")

    def overridden_method(self):
        print("Overridden Method: Overridden in class C")
#________________________________________
#Step 2: Creating a Main Class and Testing Methods
#We instantiate objects for A, B, and C, then call methods specific to each.
class Main:
    def run():
        # Creating objects
        objA = A()
        objB = B()
        objC = C()

        print("\n--- Calling Methods Using Their Own Objects ---")
        objA.method1()
        objA.method2()
        objA.overridden_method()

        objB.method3()
        objB.method4()
        objB.overridden_method()

        objC.method5()
        objC.method6()
        objC.overridden_method()

        print("\n--- Calling Overridden Method Using Superclass Reference ---")
        refA_to_B = B()
        refA_to_C = C()

        refA_to_B.overridden_method()  # Calls B's version
        refA_to_C.overridden_method()  # Calls C's version

        print("\n--- Runtime Polymorphism with Data Members ---")
        print("A's data:", objA.data)
        print("B's data:", objB.data)
        print("C's data:", objC.data)

        # Superclass reference holding subclass object
        refA_to_B = B()
        refA_to_C = C()

        print("\nAccessing data members with superclass reference:")
        print("refA_to_B.data:", refA_to_B.data)  # B's data
        print("refA_to_C.data:", refA_to_C.data)  # C's data

# Running the program
Main.run()
#________________________________________
#Explanation of Key Concepts
#1.      Inheritance:
#•       B inherits from A → Gains access to all A methods.
#•       C inherits from B → Gains access to all A and B methods.
#2.      Method Overriding:
#•       overridden_method() is overridden in both B and C.
#•       When called on a subclass object, the latest overridden version executes.
#3.      Runtime Polymorphism:
#•       overridden_method() is invoked using superclass reference.
#•       refA_to_B and refA_to_C dynamically call the subclass versions.
#4.      Polymorphism with Data Members:
#•       Each class has a data instance variable.
#•       Changing values in subclasses affects them independently.
#________________________________________
#Program Output
#--- Calling Methods Using Their Own Objects ---
#Method1: Specific to class A
#Method2: Specific to class A
#Overridden Method: Defined in class A
#Method3: Specific to class B
#Method4: Specific to class B
#Overridden Method: Overridden in class B
#Method5: Specific to class C
#Method6: Specific to class C
#Overridden Method: Overridden in class C

#--- Calling Overridden Method Using Superclass Reference ---
#Overridden Method: Overridden in class B
#Overridden Method: Overridden in class C

#--- Runtime Polymorphism with Data Members ---
#A's data: Data from A
#B's data: Data from B
#C's data: Data from C

#Accessing data members with superclass reference:
#refA_to_B.data: Data from B
#refA_to_C.data: Data from C
```

**9.Abstract Class**  

```python
# 1. Create an abstract class with abstract and non-abstract methods.
# In Python, we can create an abstract class using the ABC module from abc. Abstract classes can have abstract methods (which must be implemented by subclasses) and non-abstract methods (which can be inherited without modification).
# Example: Abstract Class with Abstract and Non-Abstract Methods
from abc import ABC, abstractmethod

# Defining an abstract class
class Vehicle(ABC):

    @abstractmethod
    def start(self):  # Abstract method
        pass

    @abstractmethod
    def stop(self):  # Another abstract method
        pass

    def description(self):  # Non-abstract method
        print("This is a vehicle.")

```

```python
# Subclass implementing the abstract methods
class Car(Vehicle):

    def start(self):
        print("Car is starting with a key.")

    def stop(self):
        print("Car is stopping with brakes.")
```

```python
# Creating an object of the subclass
car = Car()
car.start()  # Calls the implemented method in Car
car.stop()  # Calls the implemented method in Car
car.description()  # Calls the inherited non-abstract method
```

```python
# 2. Create a sub class for an abstract class. Create an object in the child class for the
# abstract class and access the non-abstract methods.

# Python program demonstrating:
#	An abstract class with both abstract and non-abstract methods.
#	A subclass that implements the abstract methods.
#	Creating an object in the child class to access the non-abstract method.

# Step 1: Creating an Abstract Class
from abc import ABC, abstractmethod

# Abstract class
class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass  # Abstract method must be overridden in subclasses

    @abstractmethod
    def move(self):
        pass  # Another abstract method

    def description(self):  # Non-abstract method
        print("Animals have unique characteristics.")

```

```python
# Step 2: Subclass Implementing Abstract Methods
class Dog(Animal):

    def sound(self):
        print("Dog barks.")

    def move(self):
        print("Dog runs.")

# Creating an object of the child class
dog = Dog()

# Accessing non-abstract method
dog.description()

# Accessing overridden abstract methods
dog.sound()
dog.move()
```

```python
# 3. Create an instance for the child class in child class and call abstract methods
# Step 1: Creating an Abstract Class
# from abc import ABC, abstractmethod

from abc import ABC, abstractmethod

# Abstract base class
class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass

# Child class
class Dog(Animal):
    def make_sound(self):
        print("Woof!")

    def __init__(self):
        print("Creating instance within the child class...")
        # Creating instance of Dog inside its own class
        self.self_instance = Dog.__new__(Dog)  # Avoid recursion by skipping __init__
        self.self_instance.make_sound()

# Instantiate Dog
dog = Dog()

```

```python
# 4.Create an instance for the child class in child class and call non-abstract methods
# Below is a Python program where:
# An abstract class contains both abstract and non-abstract methods.
#	A child class implements the abstract methods.
#	The child class creates an instance of itself and calls the non-abstract method.

class Laptop:
    def power_on(self):
        print("Laptop is powering on...")

    def power_off(self):
        print("Laptop is shutting down.")

class DellLaptop(Laptop):
    def greet_user(self):
        print("Welcome to your Dell Laptop!")

    def __init__(self):
        print("Creating DellLaptop instance inside its own class...")
        self.self_instance = DellLaptop.__new__(DellLaptop)  # Avoid recursion
        # Call non-abstract methods defined in this class
        self.self_instance.greet_user()
        # Also call inherited non-abstract methods
        self.self_instance.power_on()
        self.self_instance.power_off()

# Instantiate the child class
dell = DellLaptop()
```

```python
# 2. Call the constructors(both default and argument constructors) of super class from a child class

# In Python, constructors of a superclass can be called from a subclass using super(). Here's a demonstration of calling both the default constructor and argument constructors of a superclass from a child class.

#Step 1: Create a Superclass with Multiple Constructors
class Parent:
    def __init__(self, name="Default Name"):
        print(f"Parent Constructor: Name = {name}")

#Step 2: Create a Subclass and Call Superclass Constructors
class Child(Parent):
    def __init__(self, name="Default Child Name", age=0):
        super().__init__(name)  # Calling Parent constructor
        print(f"Child Constructor: Age = {age}")

# Creating objects to test constructor calls
print("\n--- Default Constructor Call ---")
child1 = Child()

print("\n--- Argument Constructor Call ---")
child2 = Child("kammary", 25)
```

```python
# 3. Apply private, public, protected and default access modifiers to the constructor

# In Python, access modifiers for constructors (__init__()) are not explicitly enforced like in Java or C++. However, we can simulate private, protected, public, and default access using naming conventions:

#1. Public Constructor
#	Accessible from anywhere.
class PublicExample:
    def __init__(self):
        print("Public Constructor: Accessible from anywhere.")

# Creating an instance
obj = PublicExample()

#2. Protected Constructor
#	Indicated by a single underscore (_).
# Meant to be used only within the class or subclasses.
class ProtectedExample:
    def __init__(self):
        print("Protected Constructor: Accessible within the class and subclasses.")

class Child(ProtectedExample):
    def access_parent(self):
        super().__init__()

# Creating an instance of Child class
child = Child()

#3. Private Constructor
#Indicated by double underscores (__).
#Cannot be accessed directly outside the class.
#Can be accessed via name mangling (_ClassName__method).
class PrivateExample:
    def __init__(self):
        print("Private Constructor: Restricted access.")

    @staticmethod
    def create_instance():
        return PrivateExample()  # Factory method to access private constructor

# Creating an instance indirectly
obj = PrivateExample.create_instance()

#4. Default (Implicit Public) Constructor
#	In Python, all constructors are public by default unless explicitly made private or protected.
class DefaultExample:
    def __init__(self):
        print("Default Constructor: Implicitly public.")

# Instantiating the class
obj = DefaultExample()

# Key Takeaways
#	Python does not enforce strict access control but uses naming conventions:
#	def __init__() → Public (default)
#	def _init() → Protected (used within subclasses)
#	def __init__() → Private (restricted access, uses name mangling)
#	Factory methods (@staticmethod) help access private constructors safely.
```

```python
#4. Write a program which illustrates the concept of attributes of a constructor.

#In Python, attributes of a constructor are the instance variables defined inside the __init__() method. These attributes store information specific to each object and allow for object-oriented programming.

#Example: Demonstrating Constructor Attributes
class Person:
    def __init__(self, name, age, city):
        # Attributes of the constructor
        self.name = name
        self.age = age
        self.city = city

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}, City: {self.city}")

# Creating objects with different attributes
person1 = Person("kammary", 25, "Hyderabad")
person2 = Person("laxmi", 30, "Bangalore")

# Accessing attributes through a method
person1.display_info()
person2.display_info()

# Accessing attributes directly
print(f"Person1's Age: {person1.age}")
print(f"Person2's City: {person2.city}")
```

```python
# 1. Write two methods with the same name but different number of parameters of same type and call the methods

class Example:
    def display(self, *args):  # Variable-length arguments
        if len(args) == 1:
            print(f"Single parameter: {args[0]}")
        elif len(args) == 2:
            print(f"Two parameters: {args[0]}, {args[1]}")
        else:
            print("Invalid number of parameters")
# Example usage:
obj = Example()
obj.display(10)       # Calls method with one parameter
obj.display(20, 30)   # Calls method with two parameters
```

```python
#2. Write two methods with the same name but different number of parameters of different data type and call the methods

class Example:
    def display(self, *args):
        if len(args) == 1:
            print(f"Single parameter: {args[0]}, Type: {type(args[0])}")
        elif len(args) == 2:
            print(f"Two parameters: {args[0]} (Type: {type(args[0])}), {args[1]} (Type: {type(args[1])})")
        else:
            print("Invalid number of parameters")

# Example usage:
obj = Example()
obj.display(10)         # Calls method with an integer
obj.display("Hello", 3.14)  # Calls method with a string and a float
```

```python
#3. Write two methods with the same name and same number of parameters of same type.
class Example:
    def display(self, num):
        if isinstance(num, int):
            print(f"Integer value: {num}")
        elif isinstance(num, float):
            print(f"Float value: {num}")
        else:
            print("Unsupported data type")
# Example usage:
obj = Example()
obj.display(10)      # Calls method with an integer
obj.display(3.14)    # Calls method with a float

```
